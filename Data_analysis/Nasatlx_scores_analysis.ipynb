# Nasa-TLX data analysis

This notebook analyzes cognitive load measurements using the NASA Task Load Index (NASA-TLX) ratings, comparing teacher workload when using XAI (Explainable AI) versus non-XAI scoring interfaces.

## Imports

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import glob
```

## Loading the data

```python
folder_path = "Nasatlx_data/"
excel_paths = glob.glob(folder_path + "*.xlsx")

# Create individual variables for each teacher's nasatlx data
bio_1 = pd.ExcelFile(excel_paths[0])
bio_2 = pd.ExcelFile(excel_paths[1])
bio_3 = pd.ExcelFile(excel_paths[2])
bio_4 = pd.ExcelFile(excel_paths[3])
bio_5 = pd.ExcelFile(excel_paths[4])
eco_1 = pd.ExcelFile(excel_paths[5])
eco_2 = pd.ExcelFile(excel_paths[6])
eco_3 = pd.ExcelFile(excel_paths[7])
eco_4 = pd.ExcelFile(excel_paths[8])
eco_5 = pd.ExcelFile(excel_paths[9])

# Create a dictionary mapping teacher labels to variables
excel_files = {
    "Biology 1": bio_1,
    "Biology 2": bio_2,
    "Biology 3": bio_3,
    "Biology 4": bio_4,
    "Biology 5": bio_5,
    "Economics 1": eco_1,
    "Economics 2": eco_2,
    "Economics 3": eco_3,
    "Economics 4": eco_4,
    "Economics 5": eco_5
}

ordered_by_index_excel_files = {
    "Biology 1": bio_1,
    "Economics 1": eco_1,
    "Biology 2": bio_2,
    "Economics 2": eco_2,
    "Biology 3": bio_3,
    "Economics 3": eco_3,
    "Biology 4": bio_4,
    "Economics 4": eco_4,
    "Biology 5": bio_5,
    "Economics 5": eco_5
}
```

## Inspecting data

### Available data of a single participant

Below we print the structure of an arbitrary participant (first biology participant) to visualize the data that we are working with.

```python
excel = pd.ExcelFile(bio_1)

df_pairwise = pd.read_excel(excel, sheet_name='weights', skiprows=19, usecols=[1,2, 3])
df_scores1 = pd.read_excel(excel, sheet_name='task 1 rating', skiprows=1, nrows=7, usecols=[0, 1, 2])
df_scores2 = pd.read_excel(excel, sheet_name='task 2 rating', skiprows=1, nrows=7, usecols=[0, 1, 2])

display("Table of pairwise comparison (the weighting of the dimensions)", df_pairwise)
display("Table of NASA-TLX scores for task 1 (XAI task)", df_scores1)
display("Table of NASA-TLX scores for task 2 (non-XAI task)", df_scores2)
```

### Combine data per participant group (biology and economics)

```python
# Calculate the weighted scores for each task to compare the XAI and non-XAI tasks
# The scores are weighted by the pairwise comparison values

results = []
for name, excel in excel_files.items():
    # Read score sheets for both tasks
    df_scores1 = pd.read_excel(excel, sheet_name='task 1 rating', skiprows=1, nrows=6, usecols=[0, 1, 2])
    df_scores2 = pd.read_excel(excel, sheet_name='task 2 rating', skiprows=1, nrows=6, usecols=[0, 1, 2])

    # Calculate the weighted scores
    sum_t1 = df_scores1['Gewogen score'].sum()
    sum_t2 = df_scores2['Gewogen score'].sum()

    # Save in results list
    results.append({
        "Teacher": name,
        "XAI task": sum_t1,
        "Non-XAI task": sum_t2
    })
    
df_results = pd.DataFrame(results)
bio_results = df_results[df_results['Teacher'].str.startswith('Biology')]
eco_results = df_results[df_results['Teacher'].str.startswith('Economics')]
```

## NASA-TLX profiles per participant

```python
# Display the NASA-TLX results of each task per participant with radar charts

fig = plt.figure(figsize=(10, 20))

# NASA-TLX dimensions that were included in the study
categories = ['Mental', 'Temporal', 'Performance', 'Frustration', 'Effort']
N = len(categories)

# Create angles for each axis
angles = [n / float(N) * 2 * np.pi for n in range(N)]
angles += angles[:1] 

radar_charts = {}
for i, (naam, excel) in enumerate(ordered_by_index_excel_files.items(), 1):
    ax = plt.subplot(5, 2, i, polar=True)
    ax.set_yticklabels([])
    
    # Read raw scores for both tasks
    df_scores1 = pd.read_excel(excel, sheet_name='task 1 rating', skiprows=1, nrows=6, usecols=[0, 1])
    df_scores2 = pd.read_excel(excel, sheet_name='task 2 rating', skiprows=1, nrows=6, usecols=[0, 1])

    # Take out frustration dimension
    df_scores1_without_physical_dimension = df_scores1.drop(index=1).reset_index(drop=True) 
    df_scores2_without_physical_dimension = df_scores2.drop(index=1).reset_index(drop=True)

    # Switch rows effort and frustration for better visualization
    temp_scores1_effort_row = df_scores1_without_physical_dimension.iloc[3].copy()
    df_scores1_without_physical_dimension.iloc[3] = df_scores1_without_physical_dimension.iloc[4]
    df_scores1_without_physical_dimension.iloc[4] = temp_scores1_effort_row

    temp_scores1_effort_row = df_scores2_without_physical_dimension.iloc[3].copy()
    df_scores2_without_physical_dimension.iloc[3] = df_scores2_without_physical_dimension.iloc[4]
    df_scores2_without_physical_dimension.iloc[4] = temp_scores1_effort_row
    
    # Extract raw scores
    task1_raw = df_scores1_without_physical_dimension['Ruwe score (0-100)'].values.tolist()
    task2_raw = df_scores2_without_physical_dimension['Ruwe score (0-100)'].values.tolist()
    
    # Close the loop for radar chart
    task1_raw += task1_raw[:1]
    task2_raw += task2_raw[:1]
    
    # Plot both tasks
    ax.plot(angles, task1_raw, linewidth=2, linestyle='solid', label='XAI', color='blue', alpha=0.7)
    ax.fill(angles, task1_raw, alpha=0.1, color='blue')
    
    ax.plot(angles, task2_raw, linewidth=2, linestyle='solid', label='Non-XAI', color='red', alpha=0.7)
    ax.fill(angles, task2_raw, alpha=0.1, color='red')
    
    ax.set_xticks(angles[:-1])
    ax.set_xticklabels(categories, size=8)
    ax.set_ylim(0, 100)
    
    # Add title with asterisk for Biology 1 and Biology 2
    if naam == "Biology 1" or naam == "Biology 2":
        ax.set_title(f"{naam}*", size=12, y=1.1, fontweight='bold')
    else:
        ax.set_title(naam, size=12, y=1.1, fontweight='bold')

    if i == 1:
        ax.legend(loc='upper right', bbox_to_anchor=(1.6, 1.4))

plt.tight_layout(h_pad=3.0)
plt.suptitle('NASA-TLX profiles for each participant (rated 0-100)', fontsize=16, y=1.03)
plt.show()
```

## Cognitive load comparison per task per participant

```python
# Plot the XAI task vs non-XAI task cognitive load scores for each participant

plt.figure(figsize=(10, 8))

legend_entries = {}

# First pass: plot points but build legend entries
for i, row in df_results.iterrows():
    color = 'forestgreen' if 'Biology' in row['Teacher'] else 'mediumpurple'
    marker = 'o' if 'Biology' in row['Teacher'] else 's'
    
    # Create a special label for Biology 1 and 2 with asterisks
    if row['Teacher'] in ['Biology 1', 'Biology 2']:
        label = f"{row['Teacher']}*"
    else:
        label = row['Teacher']
    
    # Add to legend entries dict to avoid duplicates
    if label not in legend_entries:
        legend_entries[label] = {'color': color, 'marker': marker}
    
    # Plot without adding to legend yet
    plt.scatter(row['XAI task'], row['Non-XAI task'], 
                s=100, color=color, marker=marker, alpha=0.7, 
                label='_nolegend_')  # Don't add to legend yet

# Add diagonal line representing equal scores
max_val = max(df_results[['XAI task', 'Non-XAI task']].max()) + 10
plt.plot([0, max_val], [0, max_val], 'k--', alpha=0.5)

# Add annotations for each point
for i, row in df_results.iterrows():
    teacher_nr = row['Teacher'].split(' ')[-1]  # Get the number from the teacher's name
    
    # Add asterisk to Biology 1 and Biology 2 in point annotations
    if row['Teacher'] in ['Biology 1', 'Biology 2']:
        annotation_text = f"{teacher_nr}*"
    else:
        annotation_text = teacher_nr
    
    plt.annotate(annotation_text, 
                (row['XAI task'], row['Non-XAI task']),
                xytext=(5, 5), textcoords='offset points')
    
plt.title('XAI vs non-XAI task cognitive load scores', fontsize=14, fontweight='bold')
plt.xlabel('XAI task cognitive load score', fontsize=12)
plt.ylabel('Non-XAI task cognitive load score', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.5)

legend_handles = []
for label, style in legend_entries.items():
    handle = plt.Line2D([0], [0], marker=style['marker'], color='w', 
                      markerfacecolor=style['color'], markersize=10, alpha=0.7, 
                      label=label)
    legend_handles.append(handle)

plt.legend(handles=legend_handles, bbox_to_anchor=(1.05, 1), 
           loc='upper left', title='Participants')

plt.tight_layout()
```

## Cognitive load per task per participant group

```python
# This plot shows the cognitive load scores for XAI and non-XAI tasks for all participants and per group

all_results = df_results
n_bio = len(bio_results)
n_eco = len(eco_results)
n_all = len(all_results)

# Create data dictionary with group information
data = {
    'Group': ['All participants', 'Biology teachers', 'Economics teachers'],
    'XAI_mean': [
        all_results['XAI task'].mean(),
        bio_results['XAI task'].mean(),
        eco_results['XAI task'].mean()
    ],
    'Non_XAI_mean': [
        all_results['Non-XAI task'].mean(),
        bio_results['Non-XAI task'].mean(),
        eco_results['Non-XAI task'].mean()
    ],
    'XAI_std': [
        all_results['XAI task'].std(),
        bio_results['XAI task'].std(),
        eco_results['XAI task'].std()
    ],
    'Non_XAI_std': [
        all_results['Non-XAI task'].std(),
        bio_results['Non-XAI task'].std(),
        eco_results['Non-XAI task'].std()
    ]
}

# Calculate standard errors from standard deviations
data['XAI_se'] = [
    data['XAI_std'][0] / np.sqrt(n_all),
    data['XAI_std'][1] / np.sqrt(n_bio),
    data['XAI_std'][2] / np.sqrt(n_eco)
]
data['Non_XAI_se'] = [
    data['Non_XAI_std'][0] / np.sqrt(n_all),
    data['Non_XAI_std'][1] / np.sqrt(n_bio),
    data['Non_XAI_std'][2] / np.sqrt(n_eco)
]

# Calculate differences
data['Difference'] = [
    data['XAI_mean'][i] - data['Non_XAI_mean'][i] 
    for i in range(len(data['Group']))
]

# Standard error of difference needs paired difference calculation
data['Difference_se'] = [
    all_results['XAI task'].sub(all_results['Non-XAI task']).std() / np.sqrt(n_all),
    bio_results['XAI task'].sub(bio_results['Non-XAI task']).std() / np.sqrt(n_bio),
    eco_results['XAI task'].sub(eco_results['Non-XAI task']).std() / np.sqrt(n_eco)
]

fig, ax = plt.subplots(figsize=(12, 7))
width = 0.35
x = np.arange(len(data['Group']))

xai_bars = ax.bar(x - width/2, data['XAI_mean'], width, 
                 label='XAI task', color='royalblue', alpha=0.8,
                 yerr=data['XAI_se'], capsize=7, ecolor='dimgray')

nonxai_bars = ax.bar(x + width/2, data['Non_XAI_mean'], width,
                    label='Non-XAI task', color='lightcoral', alpha=0.8,
                    yerr=data['Non_XAI_se'], capsize=7, ecolor='dimgray')

ax.grid(axis='y', linestyle='--', alpha=0.7)

ax.set_title('Cognitive load scores for XAI vs non-XAI tasks', fontsize=14, fontweight='bold')
ax.set_ylabel('Cognitive load score (NASA-TLX)', fontsize=12)
ax.set_xticks(x)
ax.set_xticklabels(data['Group'], fontsize=11)
ax.legend(fontsize=11)

# Add value labels on top of bars
for i, bars in enumerate(zip(xai_bars, nonxai_bars)):
    xai_bar, nonxai_bar = bars
    ax.text(xai_bar.get_x() + xai_bar.get_width()/2, xai_bar.get_height() + data['XAI_se'][i] + 1,
            f'{data["XAI_mean"][i]:.1f}', ha='center', va='bottom', fontsize=10)
    ax.text(nonxai_bar.get_x() + nonxai_bar.get_width()/2, nonxai_bar.get_height() + data['Non_XAI_se'][i] + 1,
            f'{data["Non_XAI_mean"][i]:.1f}', ha='center', va='bottom', fontsize=10)

# Add difference information with standard error
for i in range(len(data['Group'])):
    x_left = x[i] - width/2
    x_right = x[i] + width/2
    y_max = max(data['XAI_mean'][i] + data['XAI_se'][i]*2, 
                data['Non_XAI_mean'][i] + data['Non_XAI_se'][i]*2) + 5
    
    ax.plot([x_left, x_right], [y_max, y_max], 'k-')
    ax.plot([x_left, x_left], [y_max-1, y_max], 'k-')
    ax.plot([x_right, x_right], [y_max-1, y_max], 'k-')
    
    ax.text((x_left + x_right)/2, y_max + 1,
            f'Diff: {data["Difference"][i]:.2f}\n (SE = {data["Difference_se"][i]:.2f})',
            ha='center', va='bottom', fontsize=10, fontweight='bold')


plt.axhline(y=0, color='black', linestyle='-', alpha=0.3)
y_max = max([max(np.array(data['XAI_mean']) + np.array(data['XAI_se'])*2), 
             max(np.array(data['Non_XAI_mean']) + np.array(data['Non_XAI_se'])*2)]) * 1.3
ax.set_ylim(0, y_max)

plt.tight_layout()
plt.show()
```

## Average cognitive scores per dimension visualization (XAI vs. non-XAI)

```python
def extract_dimension_scores(excel_files):
    """
    Extract and organize NASA-TLX dimension scores from multiple participant Excel files.
    
    Parameters
    ----------
    excel_files : dict
        Dictionary mapping participant identifiers to pandas ExcelFile objects
        containing their NASA-TLX responses
        
    Returns
    -------
    dict
        Nested dictionary organized as:
        {
            'XAI': {
                'Mental demands': [scores for all participants],
                'Temporal demands': [scores for all participants],
                'Performance': [scores for all participants],
                'Frustration': [scores for all participants],
                'Effort': [scores for all participants]
            },
            'Non-XAI': {
                # Same structure as XAI
            }
        }
        Where each inner list contains the raw scores (0-100) for that dimension
        across all participants, in the same order as the excel_files input
    """
   
    all_scores = {
        'XAI': {'Mental demands': [], 'Temporal demands': [], 'Performance': [], 'Frustration': [], 'Effort': []},
        'Non-XAI': {'Mental demands': [], 'Temporal demands': [], 'Performance': [], 'Frustration': [], 'Effort': []}
    }
    
    # Extract scores from each participant
    for naam, excel in excel_files.items():
        # Read raw scores for both tasks
        df_scores1 = pd.read_excel(excel, sheet_name='task 1 rating', skiprows=1, nrows=6, usecols=[0, 1])
        df_scores2 = pd.read_excel(excel, sheet_name='task 2 rating', skiprows=1, nrows=6, usecols=[0, 1])
        
        # Remove physical dimension
        df_scores1 = df_scores1.drop(index=1).reset_index(drop=True) 
        df_scores2 = df_scores2.drop(index=1).reset_index(drop=True)
        
        # Assign values to categories (maintain consistent ordering)
        dimensions = ['Mental demands', 'Temporal demands', 'Performance', 'Frustration', 'Effort']
        for i, dim in enumerate(dimensions):
            row_index = i
            if i >= 3: 
                row_index = 7 - i 
            all_scores['XAI'][dim].append(df_scores1.iloc[row_index]['Ruwe score (0-100)'])
            all_scores['Non-XAI'][dim].append(df_scores2.iloc[row_index]['Ruwe score (0-100)'])
            
    return all_scores
```

```python
# Get all dimension scores
scores_by_dimension = extract_dimension_scores(ordered_by_index_excel_files)

# Convert to DataFrame for easier analysis
results = []
for dimension in scores_by_dimension['XAI'].keys():
    for i in range(len(scores_by_dimension['XAI'][dimension])):
        results.append({
            'Dimension': dimension,
            'Participant': i,
            'XAI Score': scores_by_dimension['XAI'][dimension][i],
            'Non-XAI Score': scores_by_dimension['Non-XAI'][dimension][i],
            'Difference': scores_by_dimension['XAI'][dimension][i] - scores_by_dimension['Non-XAI'][dimension][i]
        })

df_dimensions = pd.DataFrame(results)

# Calculate average scores and differences
avg_by_dimension = df_dimensions.groupby('Dimension').agg({
    'XAI Score': 'mean',
    'Non-XAI Score': 'mean',
    'Difference': ['mean', 'std']
}).reset_index()

display(avg_by_dimension)
```

```python
# Calculate standard errors for both tasks separately
n = len(df_dimensions['Participant'].unique())  # Get number of participants

def std_error(x):
    return x.std() / np.sqrt(n)

# Aggregate means, std deviations, and standard errors
std_by_dimension = df_dimensions.groupby('Dimension').agg({
    'XAI Score': ['mean', 'std', std_error],
    'Non-XAI Score': ['mean', 'std', std_error],
    'Difference': ['mean', 'std', std_error]
})

# Rename the custom function column to 'se'
std_by_dimension.columns = [
    (col[0], 'se') if col[1] == 'std_error' else col
    for col in std_by_dimension.columns
]

std_by_dimension = std_by_dimension.reset_index()

fig, ax = plt.subplots(figsize=(12, 6))

# Extract data 
dimensions = std_by_dimension['Dimension']
xai_means = std_by_dimension[('XAI Score', 'mean')]
nonxai_means = std_by_dimension[('Non-XAI Score', 'mean')]

xai_se = std_by_dimension[('XAI Score', 'se')]
nonxai_se = std_by_dimension[('Non-XAI Score', 'se')]
diff_se = std_by_dimension[('Difference', 'se')]

# Calculate difference values
difference_values = []
for i in range(len(dimensions)):
    mean_diff = xai_means[i] - nonxai_means[i]
    difference_values.append(mean_diff)

x = np.arange(len(dimensions))
width = 0.35

ax.bar(x - width/2, xai_means, width, label='XAI task', color='royalblue', alpha=0.8, 
       yerr=xai_se, ecolor='dimgray', capsize=8)
ax.bar(x + width/2, nonxai_means, width, label='Non-XAI task', color='lightcoral', alpha=0.8,
       yerr=nonxai_se, ecolor='dimgray', capsize=8)

# Add difference information for each dimension
for i in range(len(dimensions)):
    x_left = i - width/2
    x_right = i + width/2
    y_max = max(xai_means[i] + xai_se[i]*1.5, nonxai_means[i] + nonxai_se[i]*1.5) + 5
    
    ax.plot([x_left, x_right], [y_max, y_max], 'k-')
    ax.plot([x_left, x_left], [y_max-1, y_max], 'k-')
    ax.plot([x_right, x_right], [y_max-1, y_max], 'k-')
    
    ax.text((x_left + x_right)/2, y_max + 1,
            f'Diff: {difference_values[i]:.2f}\n (SE = {diff_se[i]:.2f})',
            ha='center', va='bottom', fontsize=9, fontweight='bold')
    

ax.set_title('Average NASA-TLX scores for XAI vs non-XAI tasks by dimension', fontsize=14, fontweight='bold')   
ax.set_ylabel('Average load rating (0-100)', fontsize=12, labelpad=15)
ax.set_xlabel('NASA-TLX dimensions', fontsize=12, labelpad=20) 
ax.set_xticks(x)
ax.set_xticklabels(dimensions, fontsize=11)
ax.legend(loc='upper right')
ax.grid(axis='y', linestyle='--', alpha=0.7)

y_max_all = max([max(xai_means + xai_se*2), max(nonxai_means + nonxai_se*2)]) * 1.3
ax.set_ylim(0, y_max_all)

plt.axhline(y=0, color='black', linestyle='-', alpha=0.3)

plt.tight_layout(rect=[0, 0.03, 1, 0.97])  

plt.show()
```
